```python
final_df = runs.merge(balls,on=['MATCH_ID_x','BATTER'],suffixes = ('_RUNS','_BALLS')).merge(fours,on=['MATCH_ID_x','BATTER'],how='left').merge(sixes,on=['MATCH_ID_x','BATTER'],how='left').fillna(0)
```

```python
final_df.rename(columns={
    'MATCH_ID_x':'ID',
    'BATSMAN_RUNS_RUNS':'RUNS',
    'BATSMAN_RUNS_BALLS':'BALLS',
    'BATSMAN_RUNS_x':'FOURS',
    'BATSMAN_RUNS_y':'SIXES'
},inplace=True)
```

```python
final_df['SR'] = round((final_df['RUNS']/final_df['BALLS'])*100,2)
```

```python
final_df = final_df.merge(temp_df,left_on=['ID','BATTER'],right_on=['MATCH_ID','PLAYER_NAME'],how='left').drop(columns=['PLAYER_NAME','MATCH_ID']).fillna(0)
```

```python
def dream11(row):
  score = 0

  score=score + row['RUNS'] + row['FOURS'] + 2*row['SIXES']

  if row['RUNS'] >= 100:
    score = score + 16
  elif row['RUNS'] >= 50 and row['RUNS'] < 100:
    score = score + 8
  elif row['RUNS'] >= 30 and row['RUNS'] < 50:
    score = score + 4
  elif row['RUNS'] == 0:
    score = score - 2

  if row['BALLS'] >=10:
    if row['SR'] > 170:
      score = score + 6
    elif row['SR'] > 150 and row['SR'] <= 170:
      score = score + 4
    elif row['SR'] > 130 and row['SR'] <= 150:
      score = score + 2
    elif row['SR'] > 60 and row['SR'] <= 70:
      score = score + -2
    elif row['SR'] > 50 and row['SR'] <= 60:
      score = score - 4
    elif row['SR'] <= 50:
      score = score - 6
    else:
      pass
  else:
    pass

  if row['IS_CAPTAIN'] == 1:
    score = score*2

  return score
```

```python
final_df['SCORE'] = final_df.apply(dream11,axis=1)
```

```python
export_df = final_df.sort_values('SCORE', ascending=False)[['ID','BATTER','SCORE']]
```
